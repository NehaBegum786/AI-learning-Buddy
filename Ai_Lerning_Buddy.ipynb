```python
%%writefile app.py
import streamlit as st
import google.generativeai as genai
import os

# Get API key from environment variable
GOOGLE_API_KEY = os.environ.get('GOOGLE_API_KEY')

# Configure Gemini API
genai.configure(api_key=GOOGLE_API_KEY)

model = genai.GenerativeModel("gemini-2.5-flash")

st.set_page_config(page_title="AI Learning Buddy neha", page_icon="🎓")

st.title("🎓 AI Learning Buddy neha")

topic = st.text_input("Enter a Topic")

option = st.selectbox(
    "Choose Activity",
    [
        "Explain Concept",
        "Real-Life Example",
        "Generate Quiz",
        "Ask Anything"
    ]
)

if st.button("Generate"):

    if topic == "":
        st.warning("Please enter a topic.")
    else:

        if option == "Explain Concept":
            prompt = f"Explain {topic} in simple language for a beginner."

        elif option == "Real-Life Example":
            prompt = f"Give one simple real-life example of {topic}."

        elif option == "Generate Quiz":
            prompt = f"Create 5 MCQs on {topic} with answers."

        else:
            prompt = topic

        response = model.generate_content(prompt)

        st.write(response.text)
```

```python
import os
import subprocess

# Clear nohup.out to get clean logs
subprocess.run(['bash', '-c', '> nohup.out'], capture_output=True, text=True)

# Kill any running ngrok or Streamlit processes aggressively
pkill_ngrok = subprocess.run(['pkill', '-f', 'ngrok'], capture_output=True, text=True)
killall_ngrok = subprocess.run(['killall', '-9', 'ngrok'], capture_output=True, text=True)
pkill_streamlit = subprocess.run(['pkill', '-f', 'streamlit'], capture_output=True, text=True)
pkill_python_streamlit = subprocess.run(['pkill', '-f', 'python -m streamlit'], capture_output=True, text=True)

killed_something = False
if pkill_ngrok.returncode == 0 or killall_ngrok.returncode == 0:
    print("Successfully killed existing ngrok processes.")
    killed_something = True
if pkill_streamlit.returncode == 0 or pkill_python_streamlit.returncode == 0:
    print("Successfully killed existing Streamlit processes.")
    killed_something = True

if not killed_something:
    print("No ngrok or Streamlit processes were running or could not be killed.")
    if pkill_ngrok.stderr:
        print(f"pkill ngrok stderr: {pkill_ngrok.stderr.strip()}")
    if killall_ngrok.stderr:
        print(f"killall ngrok stderr: {killall_ngrok.stderr.strip()}")
    if pkill_streamlit.stderr:
        print(f"pkill streamlit stderr: {pkill_streamlit.stderr.strip()}")
    if pkill_python_streamlit.stderr:
        print(f"pkill python -m streamlit stderr: {pkill_python_streamlit.stderr.strip()}")
```
